```python
# do some cleaning, improve the "time_to_visit" column

# List of months in order
months = ["January", "February", "March", "April", "May", "June", "July",
          "August", "September", "October", "November", "December"]

#Function
def expand_months(value):
    if not isinstance(value, str):
        return value
    
    v = value.strip().replace("\xa0", " ").lower()
    
    # Year-round
    if "year-round" in v:
        return ", ".join(months)
    
    # Month range
    if "-" in v:
        start, end = [m.strip().capitalize() for m in v.split("-")]
        if start not in months or end not in months:
            return value  # fallback if unknown month names
        start_idx = months.index(start)
        end_idx = months.index(end)
        # Handle ranges that wrap around the year
        if start_idx <= end_idx:
            return ", ".join(months[start_idx:end_idx+1])
        else:
            return ", ".join(months[start_idx:] + months[:end_idx+1])
    
    # Comma-separated months
    if "," in v:
        return ", ".join([m.strip().capitalize() for m in v.split(",")])
    
    # Single month
    return v.capitalize()

# Apply function
df["time_to_visit"] = df["time_to_visit"].apply(expand_months)
```
